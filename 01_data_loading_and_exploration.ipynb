```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
sns.set_theme(style='whitegrid', palette='muted')

print('Libraries loaded successfully ✓')
```

```python
DATA_PATH = '../data/'

files = {
    'orders'                      : 'olist_orders_dataset.csv',
    'order_items'                 : 'olist_order_items_dataset.csv',
    'order_payments'              : 'olist_order_payments_dataset.csv',
    'order_reviews'               : 'olist_order_reviews_dataset.csv',
    'customers'                   : 'olist_customers_dataset.csv',
    'sellers'                     : 'olist_sellers_dataset.csv',
    'products'                    : 'olist_products_dataset.csv',
    'geolocation'                 : 'olist_geolocation_dataset.csv',
    'product_category_translation': 'product_category_name_translation.csv'
}

dfs = {}
for alias, filename in files.items():
    filepath = os.path.join(DATA_PATH, filename)
    dfs[alias] = pd.read_csv(filepath)
    print(f'  ✓  {alias:<35} {dfs[alias].shape[0]:>8,} rows  x  {dfs[alias].shape[1]:>2} cols')

print('\nAll 9 files loaded!')
```

```python
for name, df in dfs.items():
    print(f'\n{"-"*55}')
    print(f'  TABLE: {name.upper()}')
    print(f'  Shape: {df.shape[0]:,} rows x {df.shape[1]} columns')
    print(f'{"-"*55}')
    print(df.dtypes.to_string())
```

```python
for name, df in dfs.items():
    print(f'\n{"="*60}')
    print(f'  {name.upper()}')
    print(f'{"="*60}')
    display(df.head(3))
```

```python
print(f'  {"TABLE":<35}  {"COLUMN":<40}  {"MISSING":>8}  {"(%)":>6}')
print('  ' + '-'*95)

for name, df in dfs.items():
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if missing.empty:
        print(f'  {name:<35}  {"— no missing values —":<40}')
    else:
        for col, count in missing.items():
            pct = count / len(df) * 100
            print(f'  {name:<35}  {col:<40}  {count:>8,}  {pct:>5.1f}%')
```

```python
key_tables = ['orders', 'order_items', 'products', 'order_reviews']

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, name in enumerate(key_tables):
    df = dfs[name]
    missing_pct = (df.isnull().sum() / len(df) * 100).to_frame(name='missing_%')
    sns.heatmap(
        missing_pct.T, ax=axes[i], cmap='YlOrRd',
        vmin=0, vmax=100, linewidths=0.5,
        cbar=True, annot=True, fmt='.1f'
    )
    axes[i].set_title(f'{name}  ({df.shape[0]:,} rows)', fontsize=12, fontweight='bold')
    axes[i].set_ylabel('')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=35, ha='right', fontsize=8)

plt.suptitle('Missing Values by Column (%) — Key Tables', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../outputs/missing_values_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved to outputs/missing_values_heatmap.png')
```

```python
for name in ['order_items', 'order_payments', 'order_reviews']:
    print(f'\n{"="*60}')
    print(f'  DESCRIPTIVE STATS — {name.upper()}')
    print(f'{"="*60}')
    display(dfs[name].describe())
```

```python
checks = {
    'orders'       : 'order_id',
    'customers'    : 'customer_unique_id',
    'sellers'      : 'seller_id',
    'products'     : 'product_id',
    'order_items'  : 'order_id',
    'order_reviews': 'review_id',
}

print(f'  {"TABLE":<25}  {"COLUMN":<25}  {"UNIQUE VALUES":>15}')
print('  ' + '-'*68)
for table, col in checks.items():
    n = dfs[table][col].nunique()
    print(f'  {table:<25}  {col:<25}  {n:>15,}')
```

```python
status_counts = dfs['orders']['order_status'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

status_counts.plot(kind='bar', ax=axes[0], color=sns.color_palette('muted'), edgecolor='white')
axes[0].set_title('Order Count by Status', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Order Status')
axes[0].set_ylabel('Number of Orders')
axes[0].tick_params(axis='x', rotation=30)
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height()):,}',
                     (p.get_x() + p.get_width() / 2, p.get_height()),
                     ha='center', va='bottom', fontsize=8)

top4 = status_counts.head(4)
other = status_counts.iloc[4:].sum()
pie_data = pd.concat([top4, pd.Series({'other': other})])
axes[1].pie(pie_data, labels=pie_data.index, autopct='%1.1f%%',
            colors=sns.color_palette('muted', len(pie_data)), startangle=140)
axes[1].set_title('Order Status Share (%)', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('../outputs/order_status_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved to outputs/order_status_distribution.png')
```

```python
summary_lines = ['OLIST DATASET — SUMMARY REPORT\n', '='*50 + '\n']

for name, df in dfs.items():
    summary_lines.append(f'\nTable : {name}')
    summary_lines.append(f'Shape : {df.shape[0]:,} rows x {df.shape[1]} cols')
    summary_lines.append(f'Columns: {", ".join(df.columns.tolist())}')
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        summary_lines.append(f'Missing: {dict(missing)}')
    summary_lines.append('-'*50)

with open('../outputs/dataset_summary.txt', 'w') as f:
    f.write('\n'.join(summary_lines))

print('Saved to outputs/dataset_summary.txt ✓')
```
